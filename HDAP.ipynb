#SPILITTING TARGET AND FEATURES
TARGET --> HEART DISEASE
FEATURES --> OTHERS

```python
X=heart_data.drop(columns='HeartDisease',axis=1)
Y=heart_data['HeartDisease']
```

```python
print(X)
```

```python
print(Y)
```

#SPLITTING TRAIN AND TEST DATA

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

X TRAIN =  FEATURES OF TRAINING DATA OF X
X TEST = FEATURES OF TESTING DATA OF X
Y TRAIN= TARGET OF ALL X TRAIN
Y TEST = TARGET OF X TEST

#MODEL TRAINING(LOG.REG)

```python
log_re=LogisticRegression()
```

#TRAINING LOG.REG MODEL WITH TRAINING DATASET

```python
log_re.fit(X_train,Y_train)
```

#ACCURACY SCORE
 (HOW MUCH PRECENTAGE THAT MODEL HAS BEEN PREDICTED CORRECTLY)

#ACCURACY ON TRAINING DATA

```python
X_train_prediction=log_re.predict(X_train)
training_data_accuracy=accuracy_score(X_train_prediction,Y_train)
```

```python
print(X_train_prediction)
```

```python
print(training_data_accuracy)
```

#ACCURACY FOR TEST DATA

```python
X_test_prediction=log_re.predict(X_test)
test_data_accuracy=accuracy_score(X_test_prediction,Y_test)
```

```python
print(test_data_accuracy)
```
